# 2.Top 5 Districts by Wheat Yield Increase Over the Last 5 Years

```python
from tabulate import tabulate 
mycursor.execute("""SELECT Dist_Name, MIN(WHEAT_YIELD_Kg_per_ha) AS Min_Yield, MAX(WHEAT_YIELD_Kg_per_ha) AS Max_Yield
FROM agriculture
GROUP BY Dist_Name;""")
out=mycursor.fetchall() 
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
from tabulate import tabulate 
mycursor.execute("""SELECT Dist_Name, 
       ((MAX(WHEAT_YIELD_Kg_per_ha) - MIN(WHEAT_YIELD_Kg_per_ha)) / MIN(WHEAT_YIELD_Kg_per_ha)) * 100 AS Yield_Growth_Percentage
FROM agriculture
WHERE  Year BETWEEN 2013 AND 2017
GROUP BY Dist_Name
ORDER BY Yield_Growth_Percentage DESC
LIMIT 5;
""")
out=mycursor.fetchall() 
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
from tabulate import tabulate 
mycursor.execute("""SELECT State_Name, 
       ((MAX(Oilseeds_Production_1000_tons) - MIN(Oilseeds_Production_1000_tons)) / MIN(Oilseeds_Production_1000_tons)) * 100 AS Growth_Rate_Percentage
FROM agriculture
WHERE Year BETWEEN 2013 AND 2017
GROUP BY State_Name
ORDER BY Growth_Rate_Percentage DESC
LIMIT 5;
""")
out=mycursor.fetchall() 
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
from tabulate import tabulate 
mycursor.execute("""SELECT Dist_Name, Crop, 
    (SUM(Area_Cultivated * Production) - SUM(Area_Cultivated) * SUM(Production) / COUNT(*)) / 
    (SQRT((SUM(Area_Cultivated * Area_Cultivated) - SUM(Area_Cultivated) * SUM(Area_Cultivated) / COUNT(*)) *
          (SUM(Production * Production) - SUM(Production) * SUM(Production) / COUNT(*)))) AS Correlation_Coefficient
FROM (
    SELECT Dist_Name, 'Rice' AS Crop, RICE_AREA_1000_ha AS Area_Cultivated, RICE_PRODUCTION_1000_tons AS Production FROM agriculture
    UNION ALL
    SELECT Dist_Name, 'Wheat', WHEAT_AREA_1000_ha, WHEAT_PRODUCTION_1000_tons FROM agriculture
    UNION ALL
    SELECT Dist_Name, 'Maize', MAIZE_AREA_1000_ha, MAIZE_PRODUCTION_1000_tons FROM agriculture
) AS Subquery
GROUP BY Dist_Name, Crop;""")
out=mycursor.fetchall() 
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
from tabulate import tabulate 
mycursor.execute("""SELECT Year, State_Name, 
       ((Cotton_Production_1000_tons - LAG(Cotton_Production_1000_tons) OVER (PARTITION BY State_Name ORDER BY Year)) 
       / LAG(Cotton_Production_1000_tons) OVER (PARTITION BY State_Name ORDER BY Year)) * 100 AS Growth_Percentage
FROM (
    SELECT State_Name 
    FROM agriculture
    WHERE Year = (SELECT MAX(Year) FROM agriculture)
    ORDER BY Cotton_Production_1000_tons DESC
    LIMIT 5
) AS TopStates
JOIN agriculture USING (State_Name)
ORDER BY Year, Growth_Percentage DESC;
""")
out=mycursor.fetchall() 
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
from tabulate import tabulate 
mycursor.execute("""WITH TopStates AS (
    SELECT State_Name 
    FROM agriculture
    WHERE Year = (SELECT MAX(Year) FROM agriculture)
    ORDER BY RICE_PRODUCTION_1000_tons + WHEAT_PRODUCTION_1000_tons DESC
    LIMIT 5
)
SELECT Year, State_Name, Crop, SUM(Production) AS Total_Production
FROM (
    SELECT Year, State_Name, 'Rice' AS Crop, RICE_PRODUCTION_1000_tons AS Production FROM agriculture
    UNION ALL
    SELECT Year, State_Name, 'Wheat', WHEAT_PRODUCTION_1000_tons FROM agriculture
) AS CombinedData
JOIN TopStates USING (State_Name)
WHERE Year BETWEEN 2015 AND 2024
GROUP BY Year, State_Name, Crop
ORDER BY Year, State_Name, Total_Production DESC;""")
out=mycursor.fetchall() 
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
